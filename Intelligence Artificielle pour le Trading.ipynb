## Stationarity

Stationarity est un aspect extrêmement important des séries temporelles - une grande partie de l'analyse effectuée sur des données de séries temporelles financières implique d'identifier si la série que nous voulons prédire est stationnaire, et si elle ne l'est pas, de trouver des moyens de la transformer pour qu'elle soit stationnaire.

*Moyenne d'une série temporelle $x_t$ is $E(x_t)=\mu(t)$*

*Variance d'une série temporelle $x_t$ is $\sigma^2(t)=E[(x_t - \mu(t))^2]$*

**Une série temporelle est stationnaire en moyenne si $\mu(t)=\mu$, i.e.la moyenne est constante dans le temps**

**Une série temporelle est stationnaire dans la variance si $\sigma^2(t)=\sigma^2$, i.e.la variance est constante dans le temps**

## Corrélation en série (Autocorrélation)

Un modèle de série chronologique décompose la série en trois composantes : tendance, saisonnière/cyclique et aléatoire. 

La composante aléatoire est appelée résidu ou erreur - la différence entre les valeur(s) prédite(s) et les valeur(s) observée(s). On parle d'autocorrélation lorsque ces résidus (erreurs) sont corrélés les uns aux autres. Autrement dit, si l'erreur de la $i_{th}$ dépend des erreurs de l'un des termes $0 .. i-1$ avant. Essentiellement, elle nous indique comment les observations séquentielles dans une série temporelle s'influencent mutuellement. 

Formellement, pour une série temporelle à covariance stationnaire (comme le n°3 ci-dessus, où la covariance entre les observations séquentielles n'est pas une fonction du temps), l'autocorrélation $\rho_k$ pour le retard $k$ (le nombre de pas de temps séparant deux observations séquentielles), 
$$
\rho_k = \frac{COV(x_t, x_{t - k})}{\sigma_x^2} = \frac{E[(x_t - \mu)(x_{t - k} - \mu)}{\sigma_x^2}
$$

Une valeur significative de $\rho_k$ indique que l'erreur du $i_{th}$ dépend des k termes précédents de ${i-k} .. {i-1}$.

### Pourquoi nous soucions-nous de la corrélation en série ? 

La corrélation en série est essentielle pour la validité des prédictions de notre modèle - Les résidus (erreurs) d'un TS stationnaire sont par définition non corrélés en série. Il est essentiel de tenir compte de l'autocorrélation dans notre modèle, sinon les erreurs standard de nos estimations des paramètres seront biaisées et sous-estimées, ce qui rendra invalides tous les tests que nous tenterons d'élaborer à l'aide du modèle. En termes simples, ignorer l'autocorrélation signifie que nous sommes susceptibles de tirer des conclusions incorrectes sur l'impact des variables indépendantes dans notre modèle.

## Bruit blanc

Par définition, *une série temporelle qui est un processus de bruit blanc a des erreurs non corrélées en série et la moyenne attendue de ces erreurs est égale à zéro.* Cela signifie que les erreurs (résidus) sont complètement tirées au hasard d'une certaine distribution de probabilité, c'est-à-dire qu'elles sont indépendantes et identiquement distribuées (i.i.d.). 

Si notre modèle de série chronologique produit des résidus de bruit blanc, cela signifie que nous avons réussi à capturer le processus sous-jacent et à expliquer toute forme de corrélation, ne laissant que des erreurs (résidus) complètement aléatoires. Nos valeurs prédites ne diffèrent des valeurs observées que par une composante d'erreur aléatoire qui ne peut être ni prévue ni modélisée.

La plus grande partie de l'analyse des séries temporelles consiste littéralement à essayer d'ajuster un modèle à la série temporelle de telle sorte que la série résiduelle soit indiscernable du bruit blanc.

```
import os
import sys

import pandas as pd
import numpy as np

import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
import statsmodels.api as sm
import scipy.stats as scs

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```
def tsplot(y, lags=None, figsize=(15, 10), style='bmh'):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        #mpl.rcParams['font.family'] = 'Ubuntu Mono'
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        qq_ax.set_title('QQ Plot')        
        scs.probplot(y, sparams=(y.mean(), y.std()), plot=pp_ax)

        plt.tight_layout()
    return 
```

## Marche aléatoire

Une marche aléatoire est un modèle de série temporelle où la valeur de la variable de la série temporelle augmente ou diminue (pas vers le haut ou vers le bas) avec une probabilité égale à chaque pas de temps, c'est-à-dire que la valeur attendue de l'observation actuelle est égale à l'observation précédente. Il est défini formellement ci-dessous :

$$x_t=x_{t-1}+w_t$$ , où $w_t$ est une série discrète de bruit blanc.
$$E[x_t]=x_{t-1}$$.

Cela signifie que si la TS que nous modélisons est une marche aléatoire, elle est imprévisible, car il s'agit littéralement de marches aléatoires.

La signification d'une marche aléatoire est qu'elle n'est pas stationnaire, car si la moyenne d'une marche aléatoire est toujours nulle, la covariance dépend en fait du temps. En particulier, la covariance est égale à la variance multipliée par le temps. Par conséquent, plus le temps augmente, plus la variance augmente.

Simulons une marche aléatoire 

```
# Marche aléatoire sans dérive

np.random.seed(1)
n_samples = 1000

x = w = np.random.normal(size=n_samples)
for t in range(n_samples):
    x[t] = x[t-1] + w[t]

_ = tsplot(x, lags=30)
```

## Modèles linéaires

Les modèles linéaires ou modèles de tendance représentent une TS qui peut être représentée par une ligne droite. L'équation de base est la suivante :
$$x_{t} = a_0+a_1 t + b$$$

La valeur de la série chronologique est déterminée par les coefficients, $a_0, a_1$ et *temps*. Un exemple pourrait être les ventes d'une entreprise qui augmentent du même montant à chaque pas de temps. Pour un exemple hypothétique, considérons une entreprise dont le prix de l'action était de 100$ au départ, et qui augmente de 50c à chaque pas de temps. 

```
# simuler une tendance linéaire

w = np.random.randn(250)
x = np.empty_like(w)

a0 = 100.
a1 = 0.5
for t in range(len(w)):
    x[t] = a0 + a1*t + w[t]
    
_ = tsplot(x, lags=30)  
```

## Modèles log-linéaires

Semblable aux modèles linéaires, sauf que la TS qui peut être représentée graphiquement utilise une fonction exponentielle. C'est-à-dire que la série temporelle a un taux de variation constant par rapport à chaque pas de temps. 
$$x_{t} = exp(a_0+a_1 t)$$
$$Log(x_{t}) = a_0+a_1 t$$
$$x_0 = exp(a_0)$$
$$x_t = x_{t-1} *exp(a1)$$

# Momentum Strategie

---
Le trading de momentum fait référence à une catégorie générale de stratégies qui extrapolent à partir des tendances existantes. Elles partent du principe que les actions qui montent vont continuer à monter et que les actions qui baissent vont continuer à baisser, et elles achètent et vendent en conséquence. Elles s'opposent aux stratégies de retour à la moyenne, qui s'appuient sur des tendances inverses.

Par exemple, nous voyons ci-dessous que le prix d'un actif fluctue mais qu'il est globalement en hausse.

```
# Importer des bibliothèques pour trouver la tendance linéaire et tracer les données
from statsmodels import regression
import statsmodels.api as sm
import scipy.stats as stats
import scipy.spatial.distance as distance
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import auquanToolbox.dataloader as dl
```

```
# Chargement des données de tarification pour un actif
start = '2014-01-01'
end = '2015-01-01'
data = dl.load_data_nologs('nasdaq', ['AAPL'], start, end)
prices = data['ADJ CLOSE']
dates = prices.index

# Tracer le prix de l'actif dans le temps.
plt.figure(figsize=(15,7))
plt.plot(prices['AAPL'])

# Find the line of best fit to illustrate the trend
X = np.arange(len(dates))
x = sm.add_constant(X) # Ajoutez une colonne de 1 pour que la ligne puisse avoir une ordonnée à l'origine.
model = regression.linear_model.OLS(prices['AAPL'], x).fit()
a = model.params[0] # Obtenir les coefficients de la ligne
b = model.params[1]
prices['Y_hat'] = X * b + a
plt.plot(prices['Y_hat'], 'r', alpha=0.9);
plt.ylabel('Price')
plt.legend(['AAPL', 'Trendline']);
plt.show()
```

L'hypothèse ici est que le mouvement de l'action est basé sur une certaine tendance à la hausse, capturée par le modèle linéaire, plus un certain bruit aléatoire.


## Momentum and Auto-Correlation

En finance, de nombreux éléments sont autocorrélés. Une explication intuitive de ce phénomène est que les prix sont déterminés par les négociants, qui tiennent compte des prix passés pour déterminer les nouveaux prix. La formule pourrait être

$$P_t = T_t + \epsilon$$ .

$$T_t = X + P_{1, t-1}$$ .

$$P_t = X + P_{1, t-1} + \epsilon$$ .

Où X couvre tous les autres facteurs pris en compte par les traders et $\epsilon$ est un bruit aléatoire. C'est la définition d'une série autocorrélée, dans laquelle la valeur au temps $t$ dépend des valeurs précédentes.

Certaines formes d'autocorrélation ne présentent pas de comportement momentum. Par exemple, une relation d'autocorrélation très simple pourrait être la suivante.

```
def generate_autocorrelated_values(N):
    X = np.zeros(N)
    for i in range(N-1):
        X[i+1] = X[i] + np.random.normal(0, 1)
    return X

plt.figure(figsize=(15,7))
for i in range(10):
    X = generate_autocorrelated_values(100)
    plt.plot(X)
plt.xlabel('$t$')
plt.ylabel('$X_t$');
plt.show()
```

Ces séries ne présenteront pas de comportement de momentum, car le montant des hausses et des baisses à chaque pas de temps est purement aléatoire. Cependant, si les rendements de la série sont autocorrélés, alors nous verrons un comportement de momentum. Nous pouvons simuler cela comme suit.

```
def generate_autocorrelated_values(N):
    X = np.zeros(N)
    for i in range(1, N-1):
        # Les rendements passés semblent-ils bons pour les investisseurs ?
        past_returns = X[i] - X[i-1]
        # Les investisseurs supposent que les rendements futurs seront égaux aux rendements passés et achètent à ce prix.
        X[i+1] = X[i] + past_returns + np.random.normal(0, 1)
    return X

plt.figure(figsize=(15,7))
for i in range(10):
    X = generate_autocorrelated_values(10)
    plt.plot(X)
plt.xlabel('$t$')
plt.ylabel('$X_t$');
plt.show()
```

Il est maintenant clair que si les investisseurs incorporent des informations sur les rendements passés dans le prix futur, nous verrons un comportement de momentum. Évidemment, dans les données réelles, ce comportement sera beaucoup plus bruyant et difficile à observer.

### Test pour l'Autocorrelation

Nous pouvons tester l'autocorrélation de plusieurs façons. La première consiste à vérifier la présence d'une racine unitaire. Si la série a une racine unitaire, nous devons nous inquiéter de la présence d'une autocorrélation.

```
from statsmodels.tsa.stattools import adfuller

X1 = generate_autocorrelated_values(100)
X2 = np.random.normal(0, 1, 100)

# Calculez la valeur p de la statistique de Dickey-Fuller pour tester l'hypothèse nulle selon laquelle yw a une racine unitaire.
print (X1)
_, pvalue, _, _, _, _ = adfuller(X1)
if pvalue > 0.05:
    print ("Nous ne pouvons pas rejeter l'hypothèse nulle que la série a une racine unitaire.")
else:
    print ("Nous rejetons l'hypothèse nulle que la série a une racine unitaire.")
print (X2)
_, pvalue, _, _, _, _ = adfuller(X2)
if pvalue > 0.05:
    print ("Nous ne pouvons pas rejeter l'hypothèse nulle que la série a une racine unitaire.")
else:
    print ("Nous rejetons l'hypothèse nulle que la série a une racine unitaire.")
```

## Momentum vs. Réversion moyenne

Le momentum et la réversion moyenne sont en quelque sorte des stratégies opposées. Dans un modèle momentum, on suppose que les mouvements à la hausse passés sont indicatifs de mouvements à la hausse futurs. Dans un modèle de retour à la moyenne, on suppose que les tendances haussières passées sont simplement des surévaluations temporaires et que le prix reviendra à la normale. Ces deux modèles sont importants et fonctionnent dans des cas différents.

Il est important de vérifier quel type de comportement peut être présent dans nos données avant d'essayer de développer une stratégie basée sur ce modèle.

### L'horizon temporel est important

Le comportement momentum et le comportement de retour à la moyenne peuvent tous deux se produire dans le même actif. Prenons à nouveau l'exemple d'AAPL. Nous pouvons voir une tendance à la hausse, mais il y a aussi un mouvement autour de cette tendance. Une approche pourrait être d'acheter et de conserver AAPL et de tenter de gagner de l'argent sur le long terme. Une autre approche consisterait à acheter AAPL chaque fois qu'il passe sous la ligne de tendance et à le vendre chaque fois qu'il se trouve en dessous. Cette stratégie fonctionnerait sur des périodes plus courtes.

```
# Chargement des données de tarification pour un actif
start = '2016-03-01'
end = '2017-01-01'
data = dl.load_data_nologs('nasdaq', ['AAPL'], start, end)
prices = data['ADJ CLOSE']
dates = prices.index

# Tracer le prix de l'actif dans le temps.
plt.figure(figsize=(15,7))
plt.plot(prices['AAPL'])

# Trouvez la ligne de meilleur ajustement pour illustrer la tendance.
X = np.arange(len(dates))
x = sm.add_constant(X) # Ajoutez une colonne de 1 pour que la ligne puisse avoir une ordonnée à l'origine.
model = regression.linear_model.OLS(prices['AAPL'], x).fit()
a = model.params[0] # Obtenir les coefficients de la ligne
b = model.params[1]
prices['Y_hat'] = X * b + a
plt.plot(prices['Y_hat'], 'r', alpha=0.9);
plt.ylabel('Price')
plt.legend(['AAPL', 'Trendline']);
plt.show()
```

Pour voir à quoi peut ressembler un signal de retour à la moyenne, examinons la différence entre l'actif et la prédiction. Bien sûr, cela ressemble à une bonne opportunité de retour à la moyenne, car la différence de prix traverse la ligne du zéro plusieurs fois.

```
plt.figure(figsize=(15,7))
plt.plot((prices['AAPL'] - prices['Y_hat']).values)
plt.hlines(np.mean(prices['AAPL'] - prices['Y_hat']), 0, len(dates), colors='r')
plt.hlines(np.std(prices['AAPL'] - prices['Y_hat']), 0, len(dates), colors='r', linestyles='dashed')
plt.hlines(-np.std(prices['AAPL'] - prices['Y_hat']), 0, len(dates), colors='r', linestyles='dashed')
plt.xlabel('Time')
plt.ylabel('Dollar Difference')
plt.show()
```

### L'horizon temporel est important pour la capacité prédictive

Chaque mesure de momentum sera prédictive sur des horizons temporels différents. Certaines sur des années, d'autres sur des jours. Cela sera également vrai pour différentes classes d'actifs. Certaines classes d'actifs ont un momentum sur des années et d'autres sur des jours. Il est important, lors de la construction d'un modèle de momentum, de garder à l'esprit l'horizon temporel le plus pertinent. Cet horizon temporel doit correspondre approximativement à l'horizon temporel sur lequel on rééquilibre ou quitte la transaction. Il doit également s'agir de l'horizon temporel qu'on utilise pour mesurer la capacité prédictive de notre modèle.

### IMPORTANT : Sélection des modèles

Il est important de ne pas tomber dans le piège des biais statistiques en lançant un grand nombre de modèles sur un ensemble de données et en voyant lesquels s'adaptent. Pour comprendre pourquoi cela est vrai, voir Les dangers de l'overfitting. En bref, si on lancez un grand nombre de modèles sur un ensemble de données, il est probable que l'un d'entre eux sera bon par hasard. Ce modèle n'aura pas de pouvoir prédictif sur les nouvelles données, mais uniquement sur celles qu'on a déjà vues.

Une meilleure approche consiste à laisser la création de notre modèle être guidée par des éléments sous-jacents que l'on sait être vrais sur nos données. Par exemple, voici un bon chemin de développement :

3. Testez notre hypothèse en vérifiant si ce modèle s'adapte bien à nos données.
4. Testez notre modèle hors échantillon sur de nouvelles données que l'on n'a jamais vues auparavant.

### Mesurer le momentum

Pour trader sur le momentum, nous devons d'abord être capables de le mesurer. Pour plus d'informations sur la façon de mesurer le momentum, veuillez consulter le carnet de notes Mesurer le momentum. On y trouvera différentes façons de quantifier le momentum. Nous discuterons ici de ce qu'on peut faire une fois que l'on a développé une mesure du momentum.

Tous les exemples suivants supposent qu'une bonne mesure du momentum $p$ a été développée. La notation $p(A_t)$ indiquera le momentum de l'actif $A$ et du temps $t$.

### Signaux d'entrée et de sortie

Une façon d'utiliser le momentum est d'avoir des seuils et d'entrer dans une position lorsque le momentum est significativement élevé, et de la quitter lorsque le momentum est significativement bas. Cela peut également être fait pour la vente à découvert d'actions si le momentum est dans la direction négative. Les stratégies de croisement de la moyenne mobile, présentées dans la section Mesurer le momentum, en sont un bon exemple.

### Classement des actions pour les paniers Long-Short

Une autre approche consiste à utiliser le momentum comme facteur de classement des actions. En bref, on fait le pari que les actions dont le momentum est plus élevé surpasseront celles dont le momentum est plus faible. Cette approche tend à être plus neutre par rapport au marché et plus robuste. Le momentum peut également être utilisé comme un facteur parmi d'autres dans un système de classement.

#### Momentum transversal

On parle de momentum transversal lorsque, au lieu de considérer le momentum absolu, on examine le momentum de chaque action par rapport à toutes les autres actions. Cela a l'avantage de fournir une échelle relative, plutôt qu'une mesure absolue du momentum qui n'a pas beaucoup de sens.
